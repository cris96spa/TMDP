```python
def numerical_gradient(func, mu, P_mat, xi, pi, gamma, tau, h=1e-5):
    return (func(mu, P_mat, xi, pi, gamma, tau + h) - func(mu, P_mat, xi, pi, gamma, tau - h)) / (2 * h)
```
